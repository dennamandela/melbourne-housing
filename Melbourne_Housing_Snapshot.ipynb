```python
import pandas as pd
```

```python
# membaca data dan simpan data di DataFrame
melbourne_data = pd.read_csv('melb_data.csv')

# cetak summary data melbourne
melbourne_data.describe()
```

```python
melbourne_data.columns
```

```python
# menjatuhkan nilai yang hilang (anggap na sebagai "tidak tersedia")
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
y = melbourne_data.Price
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Tentukan model. Tentukan nomor untuk random_state untuk memastikan hasil yang sama setiap kali dijalankan
melbourne_data = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_data.fit(X, y)
```

```python
print("Membuat prediksi untuk 5 rumah berikut:")
print(X.head())
print("Prediksinya adalah:")
print(melbourne_data.predict(X.head()))
```
